### torchtext

```python
from torchtext import data
```

```python
def tokenize(s):
    return s.strip().split()

bos='<s>'
eos='</s>'
pad='<blank>'
feat = data.Field(
    init_token=bos, eos_token=eos,
    pad_token=pad, tokenize=tokenize,
    include_lengths=True)

pos = data.TabularDataset(
path='data/tmp.txt', format='csv',
fields=[('text', feat)])

bos=None
eos=None
pad='<blank>'
feat_2 = data.Field(
    init_token=bos, eos_token=eos,
    pad_token=pad, tokenize=tokenize,
    include_lengths=False)

pos_2 = data.TabularDataset(
path='data/tmp.txt', format='csv',
fields=[('text', feat_2)])
```
